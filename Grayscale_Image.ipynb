```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('media\emma.jfif',cv.IMREAD_GRAYSCALE)
imgb = img + 50
imgc = cv.add(img, 50) # preferred way
f, ax = plt.subplots(1,3, figsize=(9, 3))
ax[0].imshow(img, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(imgb, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Img + 50')
ax[1].axis('off')
ax[2].imshow(imgc, cmap='gray', vmin=0, vmax=255)
ax[2].set_title('cv.add(img, 50)')
ax[2].axis('off')
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

def img_brighten_1(image, shift):
    h = image.shape[0]
    w = image.shape[1]
    result = np.zeros(image.shape, image.dtype)
    for i in range(0,h):
        for j in range(0,w):
            result[i,j] = image[i,j] + shift
    return result

def img_brighten_2(image, shift):
    h = image.shape[0]
    w = image.shape[1]
    result = np.zeros(image.shape, image.dtype)
    for i in range(0,h):
        for j in range(0,w):
            result[i,j] = min(image[i,j] + shift, 255)
    return result

img = cv.imread('media/emma.jfif',cv.IMREAD_GRAYSCALE)
imgb = img_brighten_1(img, 50)
imgc = img_brighten_2(img, 50)
f, ax = plt.subplots(1,3, figsize=(9, 3))
ax[0].imshow(img, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(imgb, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Image_brighten_1')
ax[1].axis('off')
ax[2].imshow(imgc, cmap='gray', vmin=0, vmax=255)
ax[2].set_title('Image_brighten_2')
ax[2].axis('off')
plt.show()
```

Zeroing out Green and Blut Planes

```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread('media\emma.jfif',cv.IMREAD_COLOR)
if img is None:
    print("Could not open or find the image.")  
    assert False
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Image')
plt.xticks([]), plt.yticks([])
plt.show()
img[:,:,1:3] = 0 #[all rows, all columns, set G and B channels to 0]
# 0 - Red, 1 - Green, 2 - Blue
plt.imshow(img)
plt.title('After removing G and B channels')
plt.xticks([]), plt.yticks([])
plt.show()



```
