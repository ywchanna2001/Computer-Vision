```python
import cv2 as cv
import numpy as np
import math 
hi_treshold_max = 255
title_window = "Canny"

ratio = 0.3 # low treshold/ hi treshold
hi_treshold = 255
low_treshold = math.floor(hi_treshold * ratio)

def on_trackbar(val):
    hi_treshold = cv.getTrackbarPos(trackerbar1_name, title_window)
    ratio = cv.getTrackbarPos(trackerbar2_name, title_window) / 100
    low_treshold = hi_treshold * ratio
    edges = cv.Canny(im, low_treshold, hi_treshold)
    cv.imshow(title_window, edges)

im = cv.imread(r'media/girl2.jfif', cv.IMREAD_COLOR)
if im is None:
    print("Error: Could not load image.")
    exit(0)

cv.namedWindow(title_window)
trackerbar1_name = "High treshold value"
cv.createTrackbar(trackerbar1_name, title_window, hi_treshold, hi_treshold_max, on_trackbar)
trackerbar2_name = "Low treshold %"
cv.createTrackbar(trackerbar2_name, title_window, 1, 100, on_trackbar)
on_trackbar(hi_treshold)
cv.waitKey(0)
cv.destroyAllWindows()
```
