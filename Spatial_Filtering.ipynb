Spatial Filtering(Averaging)

```python
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('media\sigiriya.jpg', cv.IMREAD_REDUCED_GRAYSCALE_2) 

kernel = np.ones((3,3),np.float32)/9
imgc = cv.filter2D(img,-1,kernel)

fig, axes = plt.subplots(1, 2, sharex='all', sharey='all', figsize=(18, 18)) 
axes [0].imshow(img, cmap='gray')
axes [0].set_title('Original Image')
axes [0].set_xticks([]), axes [0].set_yticks([])
axes [1].imshow(imgc, cmap='gray')
axes [1].set_title('Averaging Filtered Image')
axes [1].set_xticks([]), axes [1].set_yticks([])
plt.show()
```

Sobel Vertical

```python
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('media\sigiriya.jpg', cv.IMREAD_REDUCED_GRAYSCALE_2) 

# Sobel vertical
kernel = np.array([(-1, -2, -1),
                   (0, 0, 0),
                   (1, 2, 1)], dtype='float32')
imgc = cv.filter2D(img,-1,kernel)

fig, axes = plt.subplots(1, 2, sharex='all', sharey='all', figsize=(18, 18)) 
axes [0].imshow(img, cmap='gray')
axes [0].set_title('Original Image')
axes [0].set_xticks([]), axes [0].set_yticks([])
axes [1].imshow(imgc, cmap='gray', vmin=-1020, vmax=1020)
axes [1].set_title('Sobel Vertical Filtered Image')
axes [1].set_xticks([]), axes [1].set_yticks([])
plt.show()
```

Sobel Horizontal

```python
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('media\sigiriya.jpg', cv.IMREAD_REDUCED_GRAYSCALE_2) 

# Sobel vertical
kernel = np.array([(-1, 0, 1),
                   (-2, 0, 2),
                   (-1, 0, 1)], dtype='float32')
imgc = cv.filter2D(img,-1,kernel)

fig, axes = plt.subplots(1, 2, sharex='all', sharey='all', figsize=(18, 18)) 
axes [0].imshow(img, cmap='gray')
axes [0].set_title('Original Image')
axes [0].set_xticks([]), axes [0].set_yticks([])
axes [1].imshow(imgc, cmap='gray', vmin=-1020, vmax=1020)
axes [1].set_title('Sobel Vertical Filtered Image')
axes [1].set_xticks([]), axes [1].set_yticks([])
plt.show()
```
